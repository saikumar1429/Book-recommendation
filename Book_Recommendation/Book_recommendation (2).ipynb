```python
books['subtitle']=books['subtitle'].fillna("")
books['authors']=books['authors'].fillna("")
books['categories']=books['categories'].fillna("")
books['description']=books['description'].fillna("")
```

```python
books['average_rating']=books['average_rating'].fillna(books['average_rating'].mean())
books['ratings_count']=books['ratings_count'].fillna(books['ratings_count'].median())
books['num_pages']=books['num_pages'].fillna(books['num_pages'].median())
books['published_year']=books['published_year'].fillna(books['published_year'].median())

```

```python
books["thumbnail"] = books["thumbnail"].fillna("https://via.placeholder.com/150")
```

```python
books.isnull().sum()
```

```python
books=books[["title","authors","categories","description","average_rating"]]
```

```python
books
```

```python
books['description']=books['description'].str.lower()
```

```python
books["combined"] = ( books["title"] + " " +books["authors"] + " " +books["categories"] + " " +books["description"]+" "+books['average_rating'].astype(str))
```

```python
import re
```

```python
books['description']=books['description'].apply( lambda x: re.sub(r'[^a-zA-Z0-9\s]','',x))
```

```python
books['description']=books['description'].apply( lambda x: re.sub(r'\d+','',x))
```

```python
books["description"] = books["description"].apply(lambda x: " ".join(x.split()))
```

```python
books["description"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(stop_words=('english'))
vector=tfidf.fit_transform(books['combined'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
```

```python
similarity[0]
```

```python
def recomend(book):
    index=books[books["title"]==book].index[0]
    scores=similarity[index]
    books_list=sorted(list(enumerate(scores)),reverse=True,key=lambda x:x[1])[1:6]
    recommended_books = []
    for i in books_list:
        recommended_books.append(books.iloc[i[0]].title)
    
    return recommended_books
```

```python
recomend('Gilead')
```
